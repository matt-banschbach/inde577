```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load Data
penguins = sns.load_dataset('penguins')

# Verify
print(penguins.head())
```

```python
df = penguins[penguins['species'] != 'Chinstrap']
print(df.shape)
```

```python
df_clean = df.dropna()
df_clean = df_clean[['species', 'bill_length_mm', 'bill_depth_mm']]
print(df_clean.shape)

gentoo = df_clean[df_clean.species == 'Gentoo']
chinstrap = df_clean[df_clean.species == 'Adelie']

gentoo = gentoo[:100]
chinstrap = chinstrap[:100]

print(gentoo.shape)
print(chinstrap.shape)
```

```python
plt.scatter(
    gentoo.bill_length_mm, gentoo.bill_depth_mm, color="magenta", label="Gentoo"
)
plt.scatter(chinstrap.bill_length_mm, chinstrap.bill_depth_mm, color="blue", label="Adelie")
plt.xlabel('Bill Length')
plt.ylabel('Bill Depth')
plt.title('Bill Length vs Depth by Species')
plt.legend()

plt.show()
```

```python
y_species = df_clean['species'].values
y = np.where(y_species == 'Gentoo', -1.0, 1.0)
print(y)
```

```python
X = np.array(df_clean[['bill_depth_mm', 'bill_length_mm']])
# print(X.head())
# print(X.dtypes)
# print(y.dtype)
# X
```

```python
class Perceptron(object):
    """
    Single Neuron
    """
    
    def __init__(self, eta = 0.025, epochs = 50):
        self.eta = eta
        self.epochs = epochs
        
        
    def train(self, X, y):
        self.w = np.random.rand(X.shape[1])
        self.b = np.random.rand()
        self.errors_ = []
        
        for _ in range(self.epochs):
            error = 0
            for x_i, y_i in zip(X, y):
                update = self.eta * (self.predict(x_i) - y_i)
                self.w -= update*x_i
                self.b -= update
                error += update
            self.errors_.append(error)
                
                
    def net_input(self, x):
        return np.dot(x, self.w) + self.b
    
    def predict(self, x):
        return np.sign(self.net_input(x))
        
```

```python
percep = Perceptron(epochs = 1_000)
percep.train(X, y)
```

```python
y_hat = percep.predict(X)
```
